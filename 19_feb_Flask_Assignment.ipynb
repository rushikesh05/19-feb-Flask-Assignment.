###Q1. What is Flask Framework? What are the advantages of Flask Framework?

####Flask is a lightweight and flexible Python web framework that enables developers to build web applications quickly and easily. It is a micro-framework, which means it provides only the essential tools needed to build a web application, without adding unnecessary layers of complexity. Flask is based on the Werkzeug WSGI toolkit and the Jinja2 template engine.

##Some of the advantages of using Flask framework are:

* Lightweight and flexible: Flask is a lightweight framework that does not impose any specific directory structure or require a particular database or ORM. Developers have complete control over how they structure their application.

* Easy to learn: Flask is very easy to learn, even for beginners. Its simple syntax and minimalistic approach make it easy to understand and work with.

* Modular design: Flask is built around a modular design, which means developers can easily add or remove components as needed. This makes it easy to create customized applications tailored to specific needs.

* Extensible: Flask has a large and active community of developers who have created a wide range of extensions and plugins that can be used to enhance the functionality of the framework.

* Support for unit testing: Flask has built-in support for unit testing, which makes it easy to test applications as they are being developed.

* Excellent documentation: Flask has excellent documentation, with a comprehensive user guide and API reference that make it easy to get started and to troubleshoot any issues that arise.

###Q2. Create a simple Flask application to display ‘Hello World!!’.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```


###Q3. What is App routing in Flask? Why do we use app routes?

####In Flask, app routing refers to the process of mapping URLs or routes to specific functions in the application. When a user visits a specific URL, Flask uses the routing mechanism to determine which function should be called to handle the request and generate the appropriate response.

##We use app routes in Flask for several reasons:

* Handling requests: App routing allows us to define how the application should handle incoming requests based on the URL requested by the user. By defining app routes, we can map different URLs to different functions in our application, which allows us to handle various types of requests.

* Creating APIs: App routes are used to create APIs that allow other applications or services to interact with our Flask application. By defining specific URLs and mapping them to functions, we can create an API that other applications can use to access our data or functionality.

* Creating web pages: App routes allow us to create dynamic web pages by rendering templates that are specific to a given URL or route. By defining app routes that map to specific templates, we can create web pages that display dynamic content based on user input or other factors.

* Creating redirects: App routes can be used to create redirects, which allow us to direct users to a different URL or page if a particular condition is met. This can be useful for handling errors or for implementing authentication and authorization mechanisms.

####Overall, app routing is an essential component of Flask that enables developers to build flexible and dynamic web applications.

###Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.


* In Flask, the url_for() function is used for URL building. It generates a URL for a given view function based on its name and any arguments passed to it.

####Here's an example Python code that demonstrates the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
```
